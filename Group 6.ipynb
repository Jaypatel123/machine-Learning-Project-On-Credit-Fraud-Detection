```python
from sklearn.model_selection import train_test_split
X=df_train.drop(['isFraud'], axis=1).values
y=df_train['isFraud'].values
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0,test_size=0.20)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
from sklearn.metrics import roc_curve,roc_auc_score,f1_score,accuracy_score
```

```python
df_train['isFraud'].value_counts()
```

```python
sm = SMOTE(random_state=31,sampling_strategy=0.75)
X_res, y_res = sm.fit_resample(X_train, y_train)
logreg = LogisticRegression(C=0.01, solver='lbfgs',max_iter=100000)
logreg.fit(X_res,y_res)
y_pred = logreg.predict(X_test)
print("Accuracy : ",accuracy_score(y_test,y_pred))
print("Accuracy : ",roc_auc_score(y_test,y_pred))
# print(classification_report(y_test, y_pred))
# print(confusion_matrix(y_test, y_pred))
```

```python
  from sklearn.neural_network import MLPClassifier

sm = SMOTE(random_state=23,sampling_strategy=0.75)
X_res, y_res = sm.fit_resample(X_train, y_train)
clf = MLPClassifier(max_iter=100000, random_state=31).fit(X_res,y_res)
y_pred = clf.predict(X_test)
print("Accuracy : ",accuracy_score(y_test,y_pred))
print("Accuracy : ",roc_auc_score(y_test,y_pred))
# print(classification_report(y_test, y_pred))
# confusion_matrix(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
sm = SMOTE(random_state=23,sampling_strategy=0.75)
X_res, y_res = sm.fit_resample(X_train, y_train)
clf = RandomForestClassifier(max_depth=2, random_state=31).fit(X_res,y_res)
y_pred = clf.predict(X_test)
print("Accuracy : ",accuracy_score(y_test,y_pred))
print("Accuracy : ",roc_auc_score(y_test,y_pred))
```
